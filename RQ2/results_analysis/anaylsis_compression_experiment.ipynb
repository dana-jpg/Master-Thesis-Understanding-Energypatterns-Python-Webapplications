```python
# statistical analysis for compression exoeeriment (done separately because here we are comparing three versions)
from pathlib import Path
import json
import itertools
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy import stats
from matplotlib.ticker import ScalarFormatter

BASE = Path("../data/experiment_results_raw")

GZIP_PATH = BASE / "energy_results_compression_gzip.json"
IDENTITY_PATH = BASE / "energy_results_compression_identity.json"
BR_PATH = BASE / "energy_results_compression_br.json"

METRICS = {
    "Process energy per run — CPU (PROC)": "proc_energy_per_run_J",
    "Process energy per run — Package (PKG)": "pkg_energy_per_run_J",
    "Process energy per run — DRAM": "dram_energy_per_run_J",
    "System energy per run — Host": "net_sys_host_per_run_J",
    "System energy per run — Package": "net_sys_pkg_per_run_J",
    "System energy per run — DRAM": "net_sys_dram_per_run_J",
}

def mean_ci(x, alpha=0.05):
    x = np.asarray(x)
    m = x.mean()
    h = stats.sem(x) * stats.t.ppf(1 - alpha / 2, len(x) - 1)
    return m, m - h, m + h

def cliffs_delta(x, y):
    gt = sum(i > j for i in x for j in y)
    lt = sum(i < j for i in x for j in y)
    delta = (gt - lt) / (len(x) * len(y))
    mag = (
        "negligible" if abs(delta) < 0.147 else
        "small" if abs(delta) < 0.33 else
        "medium" if abs(delta) < 0.474 else
        "large"
    )
    return delta, mag

def remove_outliers(x, lower_q=0.005, upper_q=0.995):
    lo, hi = np.quantile(x, [lower_q, upper_q])
    return x[(x >= lo) & (x <= hi)]

paths = {
    "gzip": GZIP_PATH,
    "identity": IDENTITY_PATH,
    "br": BR_PATH,
}

frames = []
for variant, path in paths.items():
    with open(path) as f:
        trials = json.load(f)["trials"]
    df = pd.DataFrame(trials)
    df["variant"] = variant
    frames.append(df)

df = pd.concat(frames, ignore_index=True)

for c in df.columns:
    if c.endswith("_J"):
        df[c.replace("_J", "_mJ")] = df[c] * 1000

variants = list(paths.keys())

for label, base_col in METRICS.items():
    col = base_col.replace("_J", "_mJ")

    data = {}
    for v in variants:
        x = df[df["variant"] == v][col].dropna().to_numpy()
        x = remove_outliers(x)
        data[v] = x

    fig, ax = plt.subplots(figsize=(5.5, 4))
    ax.boxplot(
        [data[v] for v in variants],
        labels=variants,
        widths=0.6
    )

    ax.set_title(label)
    ax.set_ylabel("Energy per run (mJ)")
    ax.set_ylim(bottom=0)
    ax.grid(axis="y", alpha=0.3)

    ax.yaxis.set_major_formatter(ScalarFormatter(useMathText=False))
    ax.ticklabel_format(style="plain", axis="y")

    plt.tight_layout()
    plt.show()

    print(label)

    for v in variants:
        m, l, h = mean_ci(data[v])
        print(f"{v}: {m:.3f} mJ (95% CI [{l:.3f}, {h:.3f}])")

    print("Pairwise statistical tests:")

    for v1, v2 in itertools.combinations(variants, 2):
        x = data[v1]
        y = data[v2]

        t, p_t = stats.ttest_ind(x, y, equal_var=False)
        u, p_u = stats.mannwhitneyu(x, y, alternative="two-sided")

        pooled = np.sqrt((x.var() + y.var()) / 2)
        d = (x.mean() - y.mean()) / pooled if pooled > 0 else 0.0

        delta, mag = cliffs_delta(x, y)

        print(f"{v1} vs {v2}:")
        print(f"  Welch t-test p-value = {p_t:.3e}")
        print(f"  Mann–Whitney U p-value = {p_u:.3e}")
        print(f"  Cohen’s d = {d:.3f}")
        print(f"  Cliff’s delta = {delta:.3f} ({mag})")

    print()
```
